```python
# splitting Training dataset 
X_train=data_train.iloc[ : , :-2]
Y_train=data_train.iloc[ : ,-2]
X_train
#Y_train
```

```python
Y_train
```

```python
# splitting Testing dataset
X_test=data_test.iloc[ : , :-1]
Y_test=data_test.iloc[ : ,-1]
X_test
```

```python
Y_test
```

```python
# Feature Scaling 
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Model 1: Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
model_DTC=DecisionTreeClassifier(random_state=42)
model_DTC.fit(X_train, Y_train)
```

```python
# prediciton of Decision Tree Classifier
prediction1=model_DTC.predict(X_test)
```

```python
# Accuracy of Decision Tree Classifier
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction1)*100)

for i in range(len(prediction1)):
    print(Y_test[i], '-', prediction1[i])
```

```python
# Model 2: K-Nearest Neighbors
k_neighbors=5
from sklearn.neighbors import KNeighborsClassifier
model_knn=KNeighborsClassifier(n_neighbors=k_neighbors)
model_knn.fit(X_train, Y_train)
```

```python
# prediciton of K- Nearest Neighbors
prediction2=model_knn.predict(X_test)
```

```python
# Accuracy of K- Nearest Neighbors
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction2)*100)

for i in range(len(prediction2)):
    print(Y_test[i], '-', prediction2[i])
```

```python
# Model 3: Support Vector Machine
from sklearn.svm import SVC
model_svc=SVC(random_state=42)
model_svc.fit(X_train, Y_train)
```

```python
# prediciton of Support Vector Machine
prediction3=model_svc.predict(X_test)
```

```python
# Accuracy of Support Vector Machine
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction3)*100)

for i in range(len(prediction3)):
    print(Y_test[i], '-', prediction3[i])
```

```python
# Model 4: Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model_RFC=RandomForestClassifier(n_estimators=100, random_state=42)
model_RFC.fit(X_train, Y_train)
```

```python
# prediciton of Random Forest Classifier
prediction4=model_RFC.predict(X_test)
```

```python
# Accuracy of Random Forest Classifier
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction4)*100)

for i in range(len(prediction4)):
    print(Y_test[i], '-', prediction4[i])
```
